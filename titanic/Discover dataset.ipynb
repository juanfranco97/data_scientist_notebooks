## 6.- Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.

```python
survivors=df_titanic.loc[df_titanic['Survived'] == 1].count()
survivors=survivors.loc['Survived']
survivors
```

```python
deaths=df_titanic.loc[df_titanic['Survived'] == 0].count()
deaths = deaths.loc['Survived']
deaths
```

```python
def percentage (total, num):
    return num*100/total
```

```python
total = survivors+deaths
total
```

```python
survivors = percentage (total, survivors)
survivors
```

```python
deaths = percentage(total, deaths)
deaths
```
